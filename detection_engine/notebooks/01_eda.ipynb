# Loading the file and inspecting the structure

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the file
df = pd.read_csv("../data/raw/Tuesday-WorkingHours.pcap_ISCX.csv")

# IMPORTANT: The column names in this dataset have hidden leading spaces.
# "Flow Duration" is actually " Flow Duration" in the raw file.
# This single line fixes all of them at once.
df.columns = df.columns.str.strip()

print("Shape (rows, columns):", df.shape)
print("\nColumn names:")
for col in df.columns:
    print(" ", col)
```

# Inspect Label Column

```python
print("Unique labels in this file:")
print(df["Label"].value_counts())

print("\nAs percentages:")
print(df["Label"].value_counts(normalize=True).mul(100).round(2))
```

# Check for missing values

```python
print("Checking for null values...")
null_counts = df.isnull().sum()
null_cols = null_counts[null_counts > 0]

if len(null_cols) == 0:
    print("No null values found.")
else:
    print(f"Columns with nulls:\n{null_cols}")
```

# Check for infinity values

```python
print("Checking for infinity values...")
numeric_cols = df.select_dtypes(include=np.number).columns
inf_counts = np.isinf(df[numeric_cols]).sum()
inf_cols = inf_counts[inf_counts > 0]

if len(inf_cols) == 0:
    print("No infinity values found.")
else:
    print(f"Columns with infinity:\n{inf_cols}")
```

# Inspect rows containing infinity values

```python
inf_mask = np.isinf(df[numeric_cols]).any(axis=1)
inf_rows = df[inf_mask]

print("Rows containing infinity values:", inf_rows.shape)

print("\nLabels affected:")
print(inf_rows["Label"].value_counts())

print("\nSample rows:")
inf_rows.head()
```

# Inspect rows containing missing values

```python
nan_rows = df[df.isnull().any(axis=1)]

print("Rows containing null values:", nan_rows.shape)

print("\nLabels affected:")
print(nan_rows["Label"].value_counts())
```
